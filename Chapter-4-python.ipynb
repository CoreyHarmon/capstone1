```python
# Create a program that checks whether a number is even or odd.

# function that checks if even or odd
def even_or_odd(number):
    if number % 2 == 0:
        print ("even")
    else:
        print ("odd")


# create a function getting user input and display output in the main() function
def main():
    print("Even or Odd Checker")
    print()
    enter_integer = int(input("Enter an Integer: "))

    even_or_odd(enter_integer)

if __name__ == "__main__":
    main()
```

```python
# Create a program that converts the number of miles that you walked on a hike to the
# number of feet that you walked.


# accept a float value for the number of miles inputed


# create a function to store the title 
def display_title():
    print("Hike Calculator")
    print("=================")

# create a function to convert miles to feet, returns an int value
def convert_miles_to_feet(miles):
    feet_in_a_mile = 5280
    return int(miles * feet_in_a_mile)

# Store the code that gets
# user input and displays output in the main() function.
def main():
    display_title()
    miles_walked = float(input("How many miles did you walk?: "))
    feet_walked = convert_miles_to_feet(miles_walked)
    print(f"You walked {feet_walked} feet.")

# Call the main function to start the program
main()
```

```python
# Create a program that uses a function to simulate the roll of a die.

#!/usr/bin/env python3

import random

def roll():
    value = random.randint(1, 6)
    return value

def display_title():
    print("Dice Roller")
    print()

def roll_dice():
    die1 = roll()
    die2 = roll()
    total = die1 + die2

    print("Die 1:", die1)
    print("Die 2:", die2)
    print("Total:", total)
    if total == 2:
        print("Snake eyes!")
    elif total == 12:
        print("Boxcars!")

    print()

def main():
    display_title()
    choice = input("Roll the dice? (y/n): ")
    print()
    
    while choice == "y":
        roll_dice()
        choice = input("Roll again? (y/n): ")
        print()

# if started as the main module, call the main function
if __name__ == "__main__":
    main()


```
