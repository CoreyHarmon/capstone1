```python
#!/usr/bin/env python3

def display_title():
    print("Prime Number Checker")
    print()

def get_valid_int():
    while True:
        num = int(input("Please enter an integer between 1 and 5000: "))
        if num <= 1 or num >= 5000:
            print("Invalid integer. Please try again.")
        else:
            return num

def get_factors(num):
    factors = []
    # this loop includes 1 and num
    for i in range(1, num+1):
        remainder = num % i
        if remainder == 0:
            factors.append(i)
    return factors

def main():
    display_title()
    
    again = 'y'
    while again == 'y':
        num = get_valid_int()
        factors = get_factors(num)
        if len(factors) == 2:
            print(f"{num} is a prime number.")
        else:
            print(f"{num} is NOT a prime number.")
            print(f"It has {len(factors)} factors:", end=" ")
            for factor in factors:
                print(factor, end=" ")
            print()
        print()
        
        again = input("Try again? (y/n): ")
        print()
    print("Bye!")

# if started as the main module, call the main function
if __name__ == "__main__":
    main()
```

```python
# Create a program that keeps track of the items that a wizard can carry. 

# TODO: create display menu function
def display_menu():
    print("The Wizard Inventory program")
    print()
    print("COMMAND MENU")
    print("show - Show all items")
    print("grab - Grab an item")
    print("edit - Edit an item")
    print("drop - Drop an item")
    print("exit - Exit program")
    print()


# TODO: create SHOW function
def show(wizard_inventory):
    print("Command: show")
    for i, item in enumerate(wizard_inventory, start=1):
        print(f"{i}. {item}")
    print()


# TODO: create GRAB function (append)
def grab(wizard_inventory):
    item_name = input("Name: ")
    wizard_inventory.append(item_name)
    print(f"{item_name} was added.\n")


# TODO: create EDIT function (wizard_inventory[i] = item)
def edit(wizard_inventory):
    try:
        i = int(input("Enter the item number you want to edit: "))
        new_name = input("Enter new name: ")
        wizard_inventory[i - 1] = new_name
        print(f"Item number {i} was updated.\n")
    except (IndexError, ValueError):
        print("Error: Invalid item number.\n")


# TODO: create DROP function (remove)
def drop(wizard_inventory):
    try:
        i = int(input("Enter the item number you want to drop: "))
        dropped_item = wizard_inventory.pop(i - 1)
        print(f"{dropped_item} was dropped.\n")
    except (IndexError, ValueError):
        print("Error: Invalid item number.\n")


# TODO: create EXIT function
def exit_program():
    print("Command: exit\nBye!")


def main():
    wizard_inventory = ["wooden staff", "wizard hat", "cloth shoes"]

    while True:
        display_menu()

        command = input("Command: ").lower()

        if command == "show":
            show(wizard_inventory)
        elif command == "grab":
            grab(wizard_inventory)
        elif command == "edit":
            edit(wizard_inventory)
        elif command == "drop":
            drop(wizard_inventory)
        elif command == "exit":
            exit_program()
            break
        else:
            print("Error: Invalid command. Please enter a valid command.\n")

if __name__ == "__main__":
    main()












 
```

```python
# Create a program that a user can use to manage the primary email address and phone
# number for a contact. 



def menu():
    print("list - Display all contacts")
    print("view - View a contact ")
    print("add - Add a contact ")
    print("del - Delete a contact ")
    print("exit - Exit program ")
    print()


# user input name and number


def display(contacts):
     if len(contacts) == 0:
        print("There are no contacts in the list.\n")
        return
     else:
        for i, row in enumerate(contacts, start=1):
            print(f"{i}. {row[0]}")
        print()


def view():
    number = int(input("Number: "))
    if number < 1 or number > len(contacts):
        print("Invalid contact number")
    else:
        contact = contacts[number-1]
        print("Name:", contact[0])
        print("Email:", contact[1])
        print("Phone:", contact[2])
        print()


def add(contacts):
    name = input("Name: ")
    email = input("Email: ")
    phone = input("Phone: ")
    
    contact = []
    contact.append(name)
    contact.append(email)
    contact.append(phone)
    contacts.append(contact)
    
    print(f"{contact[0]} was added.")   
    print()


def delete(contacts):
    number = int(input("Number: "))
    if number < 1 or number > len(contacts):
        print("Invalid contact number.\n")
    else:
        contact = contacts.pop(number-1)
        print(f"{contact[0]} was deleted.\n")


# function for view and delete - display error message using print

contacts = [["Guido van Roddum", "guido@gmail.com", "+44 45 8754 0845"], ["Eric Idle", "eric@ericidle.com", "+44 20 7946 0958"]]

def main():
    menu()

while True:
    command = input("Enter command: ")
    if command == "list":
        display(contacts)
    elif command == "view":
        view(contacts)
    elif command == "add":
        add(contacts)
    elif command == "delete":
        delete(contacts)
        break
    else:
        print("Not a valid command, please try again.")

if __name__ == "__main__":
    main()

    
```
